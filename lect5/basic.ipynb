wap to which will take list as input from user and will print the largest element

```python

lst = []

n = int(input("Enter number of elements: "))

for i in range(n):
    num = int(input("Enter element: "))
    lst.append(num)

large = lst[0]

for i in lst:
    if i > large:
        large = i

print("Largest element is:", large)
```
