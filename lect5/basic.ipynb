```python

lst = []

n = int(input("Enter number of elements: "))

for i in range(n):
    num = int(input("Enter element: "))
    lst.append(num)

large = lst[0]

for i in lst:
    if i > large:
        large = i

print("Largest element is:", large)
```

```python
lst = []

n = int(input("Enter number of elements: "))

for i in range(n):
    num = int(input("Enter element: "))
    lst.append(num)

largest = second = lst[0]

for num in lst:
    if num > largest:
        second = largest
        largest = num
    elif num > second and num != largest:
        second = num

print("Largest:", largest)
print("Second Largest:", second)
```
