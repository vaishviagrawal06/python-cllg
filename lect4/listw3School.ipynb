LIST 
1. used to store multiple items in a single variable.
2. Lists are one of 4 built-in data types in Python 
3. they are used to store collections of data, the other 3 are :- 
Tuple, Set, and Dictionary
4. ordered, changeable, and allow duplicate values.

```python
thislist = ["apple","banana","cherry"]
print(thislist)
```

ORDERED 
1. that the items have a defined order, and that order will not change.

2. If you add new items to a list, the new items will be placed at the end of the list.

CHANGEABLE
1.  we can change, add, and remove items in a list after it has been created.




```python
thislist = ["apple", "banana", "cherry", "apple", "cherry"]
print(thislist)
```
