```python
# login program and indentation
# email -> hello.03@gmail.com
# password -> 1234

email = input("enter email")
password= input("enter password")

if email == 'hello.03@gmail.com' and password=='1234':
    print('welcome')
elif email == 'hello.03@gmail.com' and password!='1234':
    print('incorrect password entered')
    
    password=input("enter password again")
    if password=='1234':
        print("finally entered the correct password")
    else:
        print('tumse na ho paayega!')
else:
    print('not correct')
```

```python
# if-else examples
# 1. Find the min of 3 given numbers
# 2. Menu Driven Program
```

```python
# 1. min of 3 num

num1 = int(input('enter first number'))
num2 = int(input('enter second number'))
num3 = int(input('enter third number'))

if num1<num2 and num1<num3:
    print(num1,' is smallest')
elif num2<num1 and num2<num3:
    print(num2,' is smallest')
else:
    print(num3,' is smallest')
```

```python
# 2. menu driven program

menu = input("""
1. enter 1 for check balance
2. enter 2 for withdrawal
3. enter 3 for exist
""")
if menu =='1':
    print('check balance')
elif menu =='2':
    print("withdrawal")
else:
    print("exit")
```
