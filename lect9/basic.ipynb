```python
# remove
s = {1, 2, 3, 5, 4}
print(s)
s.remove(4)    #4 is removed from set s
print(s)
```

```python
# discard
#remove an element 
# discard me agar vo ele set me nhi hai to vo error throw nhi karega
s = {1, 2, 3, 5, 4}
print(s)
s.discard(2)
print(s)
```

```python
st={1,2,3,4,'hello',(5,6,3,7)}
st.remove(3)
print(st)
```

```python
# del keyword complete set ko delete karta hai
st = {3,4,5,6,7}
del st # This will delete whole set from the memory.
print(st) 
```

```python
# add
st={1,2,3,4,3.2,'hello',(2,7,8,5)}
st.add(20)
st.add((20,3,4))
# st.add([40,45,56]) 
# ⬆️error generated
st.update((1000,3,4))
st.update([50,73,4])
print(st)
```

```python
s={1,2,3,4}
s1={3,4,5,6,7,8}
print("union",s|s1)
print("intersection",s&s1)
print("difference",s-s1)
print("symmetric difference",s^s1)
```
