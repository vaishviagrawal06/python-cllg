```python
m = int(input("enter a month"))
if m>=3 and m<=5:
    print("spring")
elif m>=6 and m<=8:
    print("summer")
elif m>=9 and m<=11:
    print("autumn")
else:
    print("winter") 
```

```python
# wap to check if a number is divible by both 3 and 5 
num = int(input("enter a number"))
if num%3==0 and num%5==0:
    print("number is divisible by both 3 and 5 ")
else:
    print("not divible by 3 and 5")
```

```python
# wap to check whether a year a leap year or not
num = int(input("enter a year"))
if (num%4==0 and num%100==0) or (num%400==0):
    print("is leap year")
else:
    print("not a leap year")
```

```python
# wap to check if the number is positive,negative or zero 
# input : enter a number = 8
# output : positive

number = int (input("enter a number "))
if number >0 :
    print("positive")
elif number<0:
    print("negative")
else:
    print("invalid number entered")
```
