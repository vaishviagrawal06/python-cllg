============OPERATORS==========

```python
#arithmetic operator

print(1+2)

print(1-2)

print(1*2)

print(1/2)

print(1//2) # agar quotient decimal hai to use integer me convert kar deta hai

print(1%2)

print(1**2)
```

```python
# RELATIONAL OPERATOR

print(4>5)

print(4<5)

print(4>=5)

print(4<=5)

print(4==5)

print(4!=5)
```

```python
# LOGICAL OPERATOR

print(1 and 0)

print(1 or 0)

print(not 0)
```

```python
# BITWISE OPERATOR 

#bitwise and
print( 2 & 3)  
# 2 and 3 biwise and operator --> same 1 different 0
# 010 
# 110
# ---
# 010

#bitwise or
print(2 | 3)

#bitwise xor -->dono same hai to 0 otherwise 1
print(2^3)

#bitiwse not
print(~4)

#bitwise left operator
print(4>>2)

#bitwise right operator
print(5<<2)
```

```python
#Assignment operator

a=2

# a= a%2
a%=a

print(a)
```

```python
#membership operator

#in/not in

print('D' not in 'Delhi')

print(1 in[2,3,4,5,6])
```

```python
#program - find the sum of a 3 digit number entered by the user

number = int(input('enter a 3 digit number'))

a =  number%10

b = number%10

number = number//10

c = number%10

print(a+b+c)
```
