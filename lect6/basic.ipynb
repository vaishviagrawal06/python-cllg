wap which will take list as input from user and create a new list from a given list the new list have all the even number from the previous given list

```python
lst =[2,3,4,5,6,8,9,12,14,24,45]
lst1 =[]
for i in lst:
    if i%2==0:
        lst1.append(i)
print(lst1)
```

wap which will create list having all the odd numbers between 2 to 20 

```python
lst = [i for i in range(2, 20) if i % 2 != 0]
print("Odd numbers:", lst)
```
